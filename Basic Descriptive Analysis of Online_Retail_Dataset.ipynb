As we can see here minimum quantity is in negative which is logically not correct. Lets explore and treat it

```python
# Exploration
unique_count = df["Quantity"].value_counts().reset_index()
unique_count = unique_count.sort_values( by = 'index')
unique_count.head()
```

Assuming negative value happens by type error. My mistake negative sign is given.

```python
#treatment
# we treat this by converting the negative value into positive value leaving the int value same
def n_p (x):
  if x<0:
    y = abs(x)
    return y
  else:
    return x

df["Quantity"] = df["Quantity"].apply(n_p)
```

```python
#evaluation
df.describe()
```

```python
# Outlier Check for Quantity

plt.boxplot(df['Quantity'])
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Box Plot Title')
plt.show()
```

```python
# outlier check for  unit price
plt.boxplot(df['UnitPrice'])
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Box Plot Title')
plt.show()
```

```python
# Outlier treatment for Quantity
q1 = df["Quantity"].quantile(0.25)
q3 = df["Quantity"].quantile(0.75)
iqr = q3-q1
upperlimit = q3+(1.5*iqr)
lowerlimit = q1-(1.5*iqr)
#create function for condition
def limit_imputer(value):
  if value > upperlimit:
    return upperlimit
  if value < lowerlimit:
    return lowerlimit
  else:
    return value
# Apply def
df["Quantity"] = df["Quantity"].apply(limit_imputer)
```

```python
plt.boxplot(df['Quantity'])
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Box Plot Title')
plt.show()
```

```python
# Outlier treatment for unit price
# Outlier treatment for Quantity
q1 = df["UnitPrice"].quantile(0.25)
q3 = df["UnitPrice"].quantile(0.75)
iqr = q3-q1
upperlimit = q3+(1.5*iqr)
lowerlimit = q1-(1.5*iqr)
#create function for condition
def limit_imputer(value):
  if value > upperlimit:
    return upperlimit
  if value < lowerlimit:
    return lowerlimit
  else:
    return value
# Apply def
df["UnitPrice"] = df["UnitPrice"].apply(limit_imputer)
```

```python
plt.boxplot(df['UnitPrice'])
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Box Plot Title')
plt.show()
```

#Create new col of total spend

```python
df["total_spend"] = df["Quantity"] * df["UnitPrice"]
df.head()
```

Which country have highest spend?

```python
country_spend= df.groupby("Country")["total_spend"].sum().reset_index()
country_totals = country_spend.sort_values(by='total_spend', ascending=False)
```

```python
plt.figure(figsize=(15, 6))
plt.bar(country_totals['Country'], country_totals['total_spend'])
plt.xlabel('Country')
plt.ylabel('Total Spend')
plt.title('Total Spend by Country')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()
```
